```python
import numpy as np 
import pandas as pd 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df= pd.read_csv("/kaggle/input/bank-loan-dataset/Bank_loan.csv")
df.describe()
```

```python
df.count()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['Gender'].fillna(df['Gender'].mode()[0], inplace= True)
df['Married'].fillna(df['Married'].mode()[0], inplace= True)
df['Dependents'].fillna(df['Dependents'].mode()[0], inplace= True)
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0], inplace= True)
df['LoanAmount'].fillna(df['LoanAmount'].mean(), inplace= True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0], inplace= True)
df['Credit_History'].fillna(df['Credit_History'].mode()[0], inplace= True)
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le= LabelEncoder()
le_data= ['Loan_ID','Gender','Married','Education','Dependents','Self_Employed','Property_Area','Loan_Status']
for label in le_data:
    df[label]= le.fit_transform(df[label])
df.dtypes
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X= df.drop('Loan_Status', axis=1)
y= df['Loan_Status']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
ss= StandardScaler()
X_train= ss.fit_transform(X_train)
X_test= ss.fit_transform(X_test)
```

```python
X_train
```

```python
my_model= DecisionTreeClassifier()
results= my_model.fit(X_train,y_train)
```

```python
predictions= results.predict(X_test)
predictions
```

```python
from sklearn.metrics import accuracy_score 
from sklearn import metrics
from sklearn.metrics import confusion_matrix
import seaborn as sn
```

```python
model_auc= accuracy_score (y_test, predictions)
model_auc
```

```python
conf_matrix= confusion_matrix(y_test, predictions)
conf_matrix
heat_map= sn.heatmap(conf_matrix, cmap='coolwarm', annot=True)
```

```python
print("The Classification Report:","\n", metrics.classification_report(y_test, predictions))
```
